```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']    #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False      #用来正常显示负号
```

```python
data['date'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')
data.index = data.date
```

```python
data[['黄金期货价格','美元指数']].corr()
```

```python
ax = data[['黄金期货价格','美元指数']].plot(secondary_y=['美元指数'],figsize=(8, 4))
ax.set_xlabel('日期')
ax.set_ylabel('黄金期货价格')
ax.right_ax.set_ylabel('美元指数')
plt.show()
```
