```python
import pandas as pd
```

```python
data = pd.read_csv('./论坛数据.csv',usecols=['content', 'publish_time', 'view_counts'])
data.info()
```

```python
data.iloc[1]
```

```python
data = data.dropna()
data.drop_duplicates(subset=['content'], inplace=True)
data = data.reset_index(drop=True)
data.info()
```

```python
import re
import string
punctuation = string.punctuation# 定义标点符号
# 使用正则表达式替换标点符号为空格，删除特殊符号
def clean_text(text):
    text = re.sub(f"[{re.escape(punctuation)}]", " ", text)  # 将标点符号替换为空格
    text = re.sub(r"[^\w\s]", "", text)  # 删除所有非字母、非数字、非空白字符（即特殊符号）
    text = re.sub(r"\s+", " ", text)  # 将多个空格替换为单个空格
    return text.strip()
data['content'] = data['content'].apply(clean_text)
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
data['text_length'] = data['content'].astype(str).apply(len)
data = data[(data['text_length'] >= 30) & (data['text_length'] <= 1000)]
data = data.reset_index(drop=True)
data.info()
```

```python
data.iloc[:10]
```

```python
def contains_long_numbers(text):
    return bool(re.search(r'(\b\d{6,}\b(\s+\d{6,}\b)*)', text))
filtered_data = data[~data['content'].apply(contains_long_numbers)]# 过滤掉包含长串数字的行
filtered_data.info()
```

```python
filtered_data = filtered_data.dropna()
filtered_data = filtered_data.reset_index(drop=True)
filtered_data.info()
```

```python
filtered_data = filtered_data[filtered_data['content'].str.contains('黄金|金价')]
filtered_data.reset_index(drop=True, inplace=True)
filtered_data.info()
```

```python
filtered_data.to_csv('./预处理过的论坛数据.csv',index=False)
```
