```python
import pandas as pd
data_all = pd.read_csv('./Processeddata.csv',usecols=['publish_time','view_counts','summarization'])
data_all.info()
```

```python
data_4500 = data_all[:4500]
data_else = data_all[4500:]
```

##将前面处理过的数据补足

```python
data_2392 = pd.read_csv('./预处理过的2392条论坛数据.csv')
data_2392['publish_time'] = 'N'
data_2392['view_counts'] = 'N'
```

```python
data_2392
```

```python
data_4500_unique = data_4500.drop_duplicates(subset=['summarization'])
merged_data = pd.merge(data_2392, data_4500_unique[['summarization', 'publish_time', 'view_counts']], 
                       on='summarization', 
                       how='left')
```

```python
merged_data = merged_data.drop(['publish_time_x', 'view_counts_x'], axis=1)
merged_data = merged_data.rename(columns={'publish_time_y': 'publish_time', 'view_counts_y': 'view_counts'})
```

```python
merged_data.to_csv('./预处理过的2392条论坛数据.csv',index=False)
```

##处理后面的数据

```python
data_else['text_length'] = data_else['summarization'].astype(str).apply(len)
data_else = data_else[(data_else['text_length'] >= 10)]
data_else = data_else[data_else['summarization'].str.contains('黄金|金价')]
data_else = data_else.drop(columns=['text_length'])
data_else = data_else.reset_index(drop=True)
data_else.info()
```

```python
data_else.to_csv('./需要预测的论坛数据.csv',index=False)
```
