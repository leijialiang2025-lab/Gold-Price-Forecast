```python
from tensorflow.keras.models import load_model
model = load_model('text_cnn_model.h5')
model.summary()
```

```python
import numpy as np
npword = np.load('./arr.npy')
npword = npword.astype(int)
npword
```

```python
npword.shape
```

```python
x_test = npword[3638:]
```

```python
prediction = model.predict(x_test)
prediction
```

```python
for i in range(len(prediction)):
    if prediction[i,0] > 0.5:
        prediction[i,0] = 1
    else:
        prediction[i,0] = 0
```
