```python
import pandas as pd
```

```python
data = pd.read_csv('./随机森林处理过的.csv')
data
```

```python
data1 = data[data['涨跌'] != 2]
data1 = data1.reset_index(drop=True)
data1
```

```python
data2 = data1.sample(frac=1).reset_index(drop=True)
data2
```

```python
data3 = data[data['涨跌'] == 2]
data3 = data3.reset_index(drop=True)
data3
```

```python
data4 = pd.concat([data2, data3], axis=0, ignore_index=True)
data4
```

```python
data4.to_csv('./9月6日text处理前的数据.csv',index=False)
```
