```python
import pandas as pd
```

```python
data = pd.read_csv('./分词过的全部数据.csv')
data
```

```python
data.loc[50650]
```

```python
a = 2135
b = 2035
begin = b
from IPython.display import clear_output
batch_size = 10  # 每10次操作刷新一次输出
for i in range(begin, 2827): 
    if data.loc[i, '涨跌'] == 0 and data.loc[i, '相关性'] == 'N': 
        pd.set_option('display.max_colwidth', None) 
        print(data.iloc[i]) 
        bi = int(input('相关性：')) 
        if bi == 3: 
            begin = i 
            print('STOP')
            break 
        data.loc[i, '相关性'] = bi         
        # 仅每10次迭代清除输出
        if i % batch_size == 0:
            clear_output(wait=True)
    elif data.loc[i, '涨跌'] == 1 and data.loc[i, '相关性'] == 'N':
        data.loc[i, '相关性'] = 1
    # 只在处理到a行时保存数据
    if i == a: 
        data.to_csv('./分词过的全部数据.csv', index=False)
```

```python
shape = data.shape[0] 
a = 100600 
b = 100550 
begin = b 
from IPython.display import clear_output 
batch_size = 10 # 每10次操作刷新一次输出 
for i in range(begin, shape): 
    if data.loc[i, '涨跌'] == 2 and i <= a - 1: 
        pd.set_option('display.max_colwidth', None) 
        print(data.iloc[i]) 
        bi = int(input('看涨or看跌（0/1）：')) 
        if bi == 0: 
            bi2 = int(input('相关性：')) 
            data.loc[i, '相关性'] = bi2 
        elif bi ==1: 
            data.loc[i, '相关性'] = 1 
        elif bi == 3: 
            begin = i 
            print('STOP') 
            break 
        data.loc[i, '涨跌'] = bi

    if i % batch_size == 0:    # 仅每10次迭代清除输出
        clear_output(wait=True)
if i == a: 
    data.to_csv('./分词过的全部数据.csv', index=False)
```
