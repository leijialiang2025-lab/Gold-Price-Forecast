```python
a = 608
b = 500
begin = b
from IPython.display import clear_output
batch_size = 10  # 每10次操作刷新一次输出
for i in range(begin, 608): 
    if data.loc[i, 'X'] == 2 and i <= a - 1: 
        pd.set_option('display.max_colwidth', None) 
        print(data.iloc[i]) 
        bi = int(input('看涨or看跌（0/1）：')) 
        if bi == 3: 
            begin = i 
            print('STOP')
            break 
        data.loc[i, 'X'] = bi         
        # 仅每10次迭代清除输出
        if i % batch_size == 0:
            clear_output(wait=True)
    # 只在处理到a行时保存数据
    if i == a: 
        data.to_csv('./预处理过的608条论坛数据.csv', index=False)
```
